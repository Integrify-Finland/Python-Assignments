### 2. Distance Metrics are important to calculate the similarity between the vectors. They are used in both supervised and unsupervised machine learning problems. Given the points (1,2,3) and (4,5,6), write a python conditional statement to print the distance for the order 1,2,3,4. If the order is 1 then it is L1 norm(Manhattan distance), order 2 is L2 norm(Euclidean distance) and the generalized distance is given by Minkowski equation.

```python
import math 

points_1 = (1,2,3)
points_2 = (4,5,6)

p1 = [i for i in points_1]
p2 = [i for i in points_2]

order = int(input("Give the order (1,2,3,4 or higher): "))

if order == 1:    
    x = list(b - a for a, b in zip(p1, p2))
    print("L1 Norm (Manhattan distance) : {:.2f}".format(sum(list(map(abs, x)))))
    
elif order == 2:  
    print("L2 Norm (Euclidean distance) : {:.2f}".format(math.sqrt(sum((b - a)**2 for a, b in zip(p1, p2)))))

elif order == 3: 
     print("L3 Norm : {:.2f}".format((sum((b - a)**3 for a, b in zip(p1, p2)))**(1/3)))

elif order == 4: 
     print("L4 Norm : {:.2f}".format((sum((b - a)**4 for a, b in zip(p1, p2)))**(1/4)))

else:
    if order > 4:
        x = list(abs(b - a)**order for a, b in zip(p1, p2))
        print("L{m} Norm : {n:.2f}".format(m=order, n=(sum(x)**(1/order))))
```

## 1. Understand how sorting algorithms work(bubble,selection,insertion). Implement them of your own. Also compare time complexities.

## Selection sorting algorithms


```python
def selection_sort(arrayy):    
    for i in range(len(arrayy)):
        min_index = i
        for j in range(i + 1, len(arrayy)):
            if arrayy[j] < arrayy[min_index]:
                min_index = j

        arrayy[i], arrayy[min_index] = arrayy[min_index], arrayy[i]

    return arrayy
```

```python
selection_sort([7,3,2,8,1])
```

## Insertion sorting algorithms



```python
def insertion_sort(arrayy):
    for i in range(1, len(arrayy)):        
        key = arrayy[i]
        for j in range(i-1, -1, -1):
            if arrayy[j] > key:
                arrayy[j+1] = arrayy[j]       
            else:
                break      
        else:
            j = -1
        arrayy[j+1] = key
    return arrayy
```

```python
insertion_sort([23,78,45,8,32,56])
```

## bubble sorting algorithms



```python
def bubble_sort(arrayy):
    n = len(arrayy)
    
    for i in range(n):        
        for j in range(0, n-i-1):            
            if arrayy[j] > arrayy[j+1] :
                arrayy[j], arrayy[j+1] = arrayy[j+1], arrayy[j]
                
    return arrayy


bubble_sort([23,28,45,8,32,56])
```

## 3. Write a function factorial of a given number

```python
# procedure 1

def factorial(n):
    """Calculate the factorial of a given integer.
    
    Args:
        n (int): The integer to calculate the factorial of.
    
    Returns:
        int: The factorial of the given integer.
    """
    facto = 1
    for i in range(n):
        facto = facto * (i + 1)
    return facto

ask = int(input('Enter a number: '))
print('Factorial of {}! is {}'.format(ask, factorial(ask)))
```

```python
# procedure 2

def factorial(n):
    """Calculate the factorial of a given integer.
    
    Args:
        n (int): The integer to calculate the factorial of.
    
    Returns:
        int: The factorial of the given integer.
    """
    for i in range(n + 1):
        if n == 0:
            facto = 1
            return facto

        else:
            facto = 1
            for i in range(1, n + 1):
                facto = facto * i
            return facto


ask = int(input('Enter a number: '))
print('Factorial of {}! is {}'.format(ask, factorial(ask)))
```

## 4.Write a function that takes two arguments: activation_function name and value that returns the output of specific activation_function. Implement for sigmoid, binary step function, relu, leaky rely, tanh.

#### Eg : activation_function('sigmoid',0) should return 0.5 

```python
import math

def activation_function(name, value):  
    """
    Compute the value of the specified activation function for the given input value.
    
    Args:
    name (str): Name of the activation function. Possible values: "sigmoid", "binary step function", "relu", "leaky relu", "tanh".
    value (float): Input value to the activation function.
    
    Returns:
    float: Value of the activation function for the given input value.
    """
    
    if name=="sigmoid" :
        sigmoid=1 / (1 + math.exp(-value))
        return sigmoid
    
    elif name=="binary step function":
        threshold=0
        if value < 0:
            return threshold
        elif value >= 0:
            return threshold
        
    elif name=="relu":
        if value < 0:
            return 0
        elif value >= 0:
            return value
        
    
    elif name=="leaky rely":
        positive_constant=0.1
        if value < 0:
            return positive_constant*value
        elif value >= 0:
            return value
        
    elif name=="tanh":  
        tanh = (math.exp(value) - math.exp(-value)) / (math.exp(value) + math.exp(-value))    
        return tanh
```

## 1. Write a function that returns permutation. Permutation is the number of ways to arrange 'r' objects form a set of 'n' objects where order matters and the formula is
P(n, r) = n! / (n - r)!. Use the factorial function from above exercise.

```python
import math 

def permutation(n, r):
    """
    Calculate the number of permutations of r items from a set of n distinct items.

    Args:
    n (int): The number of distinct items in the set.
    r (int): The number of items to be selected for the permutation.

    Returns:
    int: The number of permutations of r items from a set of n distinct items.
    """
    
    permutation = math.factorial(n) / math.factorial(n - r)
    return permutation

permutation(5, 2)
```

## 1. Write a function that takes three arguments. First arguments should determine whether to return permutation or combination or both. Make the first argument optional and default is both if nothing is passed while calling the function.The second argument is n, which is optional and default is 10. Finally the third argument is r, which is also optional and default is 5.

```python
import math

def permutation_combination(x="both",  n=10, r=5):  
    """
    Calculate the number of permutations and combinations of r items from a set of n distinct items.

    Args:
        x (str): Indicates whether to calculate "permutation", "combination", or both "both". Defaults to "both".
        n (int): The number of distinct items in the set. Defaults to 10.
        r (int): The number of items to be selected for the permutation or combination. Defaults to 5.

    Returns:
        int or tuple of ints: The number of permutations or combinations of r items from a set of n distinct items, depending on the value of x.
        If x is "both", a tuple containing both the permutation and combination values is returned.
    """
    permutation = math.factorial(n) / math.factorial(n - r)
    combination = math.factorial(n) /  (math.factorial(r)*math.factorial(n - r))
    
    if x == "permutation":
        return permutation 
    elif x=="combination":
        return combination
    else:
        return (permutation, combination)  
```

```python
# calling permutation
permutation_combination(x='permutation', n=5, r=3)
```

```python
# calling combination
permutation_combination(x='combination', n=5, r=3)
```

```python
# calling both

permutation_combination()
```
