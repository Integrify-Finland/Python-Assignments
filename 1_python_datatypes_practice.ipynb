```python
# data type using object literals
age = 28
average = 15.82
name = "Monali"
number_list = [1,2,3,4]
data = ('a', 'b', 'c')
emp_info = {'name': "John", 'id': '312', 'salary': 40000}
ids = {'E1', 'E2', 'E4'}
print(f"Age = {age} and data type= {type(age)}")
print(f"Average = {average} and data type= {type(average)}")
print(f"Name = {name} and data type= {type(name)}")
print(f"Number list = {number_list} and data type= {type(number_list)}")
print(f"Data = {data} and data type= {type(data)}")
print(f"Employee info = {emp_info} and data type= {type(emp_info)}")
print(f"IDs = {ids} and data type= {type(ids)}")

print("-" * 50)
if isinstance(ids, set):
    print("Datatype of 'data' is Set")
else:
    print("Datatype of 'data' is not Set")

# magic methods
print("-" * 50)
print(dir(age))
if age.__gt__(20):
    print('\nQualified for the job')
    
print("-" * 50)
print(dir(number_list))
print(f"Length of list = {number_list.__len__()}")

print("-" * 50)
print(dir(emp_info))
print(f"Item from dictionary = {emp_info.__getitem__('name')}")
```

```python
# format strings
print("1 book costs $%f. %s need %d books which costs $%f"%(7.25, 'John', 5, 7.25*5))
print("-"*50)

print("1 book costs ${}. {} need {} books which costs ${}".format(7.25, 'John', 5, 7.25*5))
print("1 book costs ${0}. {1} need {3} books which costs ${2}".format(7.25, 'John', 7.25*6, 6))
print("1 book costs ${0!s}. {1} need {2!r} books which costs ${3!a}".format(7.25, 'John', 5, 7.25*5))
print("-"*50)
print("1 book costs ${cost}. {name} need {quantity} books which costs ${cost} * {quantity}".format(cost=7.45, name='John', quantity=5))
print("1 book costs ${cost}. {name} need {quantity} books which costs ${total_cost}".format(cost=7.45, name='John', quantity=5, total_cost=7.25*5))
print("-"*50)
print("1 book costs ${:f}. {:s} need {:d} books which costs ${:F}".format(7.25, 'John', 5, 7.25*5))
```

```python
# Familirizing with list, dict, tupe and set

# list excercises`
countries= ["Finland","Sweden","Denmark","Netherlands","Germany","Norway","Iceland"]
# Find the index of "Sweden"
print(f"Index of 'Sweden' \n {countries.index('Sweden')}")

# Get all the items except the first
print(f"All items except first are \n {countries[1:]}")

# Get all the items except the last
print(f"All items except last are \n {countries[:-2]}")

# Find the last item using negative indexing
print(f"Last item is \n {countries[-1]}")

# Find the last three item using negative indexing
print(f"Last three items are \n {countries[-3:]}")

# Grab all the items with odd indices
print(f"All items with odd indices are \n {countries[::2]}")

# Find the first and last item for the list
print(f"First item of the list is \n {countries[0]}")
print(f"Last item of the list is \n {countries[-1]}")

# Get [”Netherlands”,”Germany”,”Norway”] using negative indexing
print(f"Few countries using negative indexing \n {countries[-4:-1]}")
print("-"*50)

# Reverse the list
print(f"Reversed list is \n {countries[::-1]}")
countries.reverse()
print(f"Reversed list is \n {countries}")
print("-"*50)

# What is the output of countries[::-2]
#reverse the list again to return to original state
countries.reverse()
print(f"Output of countries[::-2] is \n {countries[::-2]}")

# What is the output of countries[1:100:2]?
print(f"Output of countries[1:100:2] is \n {countries[1:100:2]}")
print("-"*50)

# What is the output of countries[len(countries)-1:0:2]?
countries[len(countries)-1:0:2]
print(f"Output of countries[len(countries)-1:0:2] is \n {countries[len(countries)-1:0:2]}")

# Get the output ['Germany', 'Netherlands', 'Denmark']
reversed_list = countries[::-1]
print(f"Output is  \n {reversed_list[2:4]}")
print("-"*50)

# Use enumerate method to get index and values in the list
for idx, item in enumerate(countries):
    print(f"Index of {item} is {idx}")

```

```python
# Create dict using two lists of same length
l1 = ["a","b","c","d"]
l2 = [1,2,3,4]
print(dict(zip(l1, l2)))

# Create two dictionaries, merge them into one.
# Hint: update method/dictionary unpacking
d1 = {'a': 10, 'b':20, 'c':30}
d2 = {'x':100, 'y': 200}
d1.update(d2)
print(d1)

# Given the list of students, assign them the course.
students = ["Duy", "Laxmi","Antonio","Maria"]
course = ["Python","Database","ML"] 
students_profile = {key: course for key in students}
print(students_profile)

# Modify Antonio’s course to ["JavaScript","Database","NodeJS"]
students_profile["Antonio"] = ["JavaScript","Database","NodeJS"]
print("Dict after update")
print(students_profile)

# Add “ReactJs” to Anonio’s course
students_profile["Antonio"].append("ReactJs")
print("Dict after adding skill")
print(students_profile)

# Replace Antonio’s “ReactJs” to “Vuejs”
students_profile["Antonio"].remove("ReactJs")
students_profile["Antonio"].append("Vuejs")
print("Dict after relacing skill")
print(students_profile)

# Remove an item from dictionary
students_profile.pop("Antonio")
print("Dict after removing an item")
print(students_profile)
```
