**Exercise 2.** Everything in python is class and object. Let's verify that. Run the following code:



```python
a = 5
b = "some string"
print(type(a), type(b))

def some_function():
    return

class SomeClass:
    pass

some_list = [int, str, tuple, dict, list, print, some_function, SomeClass]
for item in some_list:
    print(type(item))
```

**Answer**<br>
The first loop (on *some_list*) prints the type of each element

**Exercise 3.** Create two classes Pet and Dog 1. Requirements for Pet class
 1. Three class attributes pet_population , pet_instance and current_year
 2. Init constructor __init__(self,name,category,age), name is name of pet,category could be dog,cat,bird etc. and age is age of pet
 3. An instance method hellofrompet that prints hello from petname
 4. Create a class method timepassedby which takes years as argumentshould automatically increase the age of all the pet object as soon as the method is applied

pet_population should show the number of pets created and pet_instance is the list of pet object<br>

Requirements for Dog class 
  1. Dog class should inherit Pet class attributes and methods. Apart from that the init constructor takes addition "breed" argument 
  2. Define the instance method barks which displays the breed does woof woof

```python
class Pet:
    """
    A class representing a pet.

    Class Attributes:
    - pet_population (int): the number of 'Pet' objects created.
    - pet_instance (list): a list of 'Pet' objects created.
    - current_year (int): the current year (default 2023).

    Attributes:
    - name (str): the name of the pet.
    - category (str): the category of the pet (e.g. dog, cat, bird, etc.).
    - age (int): the age of the pet in years.

    Methods:
    - hello_from_pet(): Prints a message indicating that the pet is saying hello.
    - time_passed_by(years): Increases the age of all 'Pet' objects by the specified number of years.
    """
    pet_population = 0
    pet_instances = []
    current_year = 2023
    
    def __init__(self, name, category, age):
        self.name = name
        self.category = category
        self.age = age
        Pet.pet_population += 1
        Pet.pet_instances.append(self)
    
    def hello_from_pet(self):
        print("Hello from " + self.name)
    
    def time_passed_by(years):
        Pet.current_year += years
        for item in Pet.pet_instances:
            item.age += years

class Dog(Pet):
    """
    A class representing a dog, which inherits from Pet.
    
    Attributes:
    - breed (str): the breed of the dog.

    Methods:
    - barks(): Prints a message indicating that the dog is barking with its breed.
    """    
    
    def __init__(self, name, category, age, breed):
        super().__init__(name, category, age)
        self.breed = breed
    
    def barks(self):
        print(self.breed + " does 'woof woof'")
```
