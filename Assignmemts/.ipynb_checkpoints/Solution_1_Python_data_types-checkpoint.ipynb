#### 3. Once you create an object of each data type, run **dir(object)**. Navigate the **magic methods/dunder methods** and other methods and attributes of each object. Try to implement those methods and attributes as many as possible.
Eg:  <br>
  a=int(5)<br>
  dir(a)<br>
  print(a.\_\_add\_\_(5))<br>
Implementing above dunder method is same as print(a+5)<br>
Verify the data types using methods **type** and **isinstance**.

```python
# Create a series of variables of different types 
var_list = [-5, 5.1, 'hello', (1,2,4,8)]

# Inspect their dunder methods:
for var in var_list:
    print("*"*103)
    print("Looping over all dunder methods of type " + str(type(var)) + " and inspecting their behaviors and output types:")
    print("*"*103)
    for dunder_method_name in dir(var):
        # Get the type as a string; e.g., if the type is "<class 'bool'>", then var_type will be "bool"
        var_type = str(type(var)).split("'")[1].replace(">","")        
        # Try to execute the dunder method - first with an input; if that fails, then try without an input
        try:
            # Try call the dunder method with 1 input (with input the same as the variable value itself)
            result = getattr(var, dunder_method_name)(var)
            print(var_type + "(" + str(var) + ")." + dunder_method_name + "(" + str(var) + ") = " + str(result) + "   (type = " + str(type(result)) + ")")
        except:
            try:
                # Try call the dunder method without input but with parentheses
                result = getattr(var, dunder_method_name)()
                print(var_type + "(" + str(var) + ")." + dunder_method_name + "() = " + str(result) + "   (type = " + str(type(result)) + ")")
            except:
                try:
                    # If that fails, try call the method without any inputs and no parentheses
                    result = getattr(var, dunder_method_name)
                    print(var_type + "(" + str(var) + ")." + dunder_method_name + " = " + str(result) + "   (type = " + str(type(result)) + ")")
                except:
                    # If that also fails, give up
                    print("Dunder method '" + dunder_method_name + "' could not be called with 0 or 1 inputes...")
        print("\n\n")

# Really cool that Python allows you to provide your own functions for operators like '+' and '*'! I didn't know that.
```

### 5. Familiarize yourself with indexing in python. 
Iterables can be accessed using index values in python. Eg: list, tuple,dict, set <br>
**a.** Create a list of items. Eg:countries= [“Finland”,“Sweden”,”Denmark”,”Netherlands”,”Germany”,”Norway”,”Iceland”] <br>
**b.** Find the index of “Sweden”. <br>
**c.** Get all the items except the first<br>
**d.** Get all the items except the last<br>
**e.** Find the last item using negative indexing<br>
**f.** Find the last three item using negative indexing<br>
**g.** Grab all the items with odd indices<br>
**h.** Find the first and last item for the list<br>
**i.** Get [”Netherlands”,”Germany”,”Norway”] using negative indexing<br>
**j.** Reverse the list<br>
**k.** What is the output of countries[::-2]<br>
**l.** What is the output of countries[1:100:2]?<br>
**m.** What is the output of countries[len(countries)-1:0:2]?<br>
**n.** Get the output ['Germany', 'Netherlands', 'Denmark']<br>

```python
# Create a list of items
countries = ["Finland", "Sweden", "Denmark", "Netherlands", "Germany", "Norway", "Iceland"] 
print("countries = " + str(countries) + "\n")

# Find the (first) index of "Sweden"
print("Index of Sweden = " + str(countries.index("Sweden"))) 

# Get all items except the first
print("All items except the first = " + str(countries[1:]))

# Get all items except the last
print("All items except the last = " + str(countries[:-1]))

# Find the last item using negative indexing
print("Last item = " + countries[-1])

# Find the last three item using negative indexing
print("Last three items = " + str(countries[-3:]))

# Grab all the items with odd indices
print("All items with odd indices = " + str(countries[1::2]))

# Find the first and last item for the list 
print("First and last item = " + str([countries[0], countries[-1]]))
print("First and last item = " + str(countries[0::len(countries)-1]))
# Note: there must be an easier way? E.g. something like countries([1 end]), as it would be done in Matlab? Possibly only with numpy?

# Get [”Netherlands”,”Germany”,”Norway”] using negative indexing
print(str(countries[-4:-1]))
    
# Reverse the list
countries.reverse()
print(countries)

# What is the output of countries[::-2]
print(countries[::-2])
# -> this grabs the items at indices -1, -3, -5, ..., 0 (start at the end of list because of the negative sign; then jump in steps of 2)

# What is the output of countries[1:100:2]?
print("countries[1:100:2]) = " + str(countries[1:100:2]))
# -> this iterates through the list from the second item up to the 100th item [or the end of the list, whichever comes first], in steps of 2
# -> in this case it will thus return ['Norway', 'Netherlands', 'Sweden'] (remember that we reversed this list)

# What is the output of countries[len(countries)-1:0:2]?
print(countries[len(countries)-1:0:2])
# -> empty, because the start index is greater than the end index while a positive step size is specified (it would give an output had we used a negative step size)

# Get the output ['Germany', 'Netherlands', 'Denmark']
output = [i for i in countries if 'e' in i and 'a' in i and 'r' in i]  
# ^---- Works, but probably not what you had in mind :-P Here is another way to do this:
print(countries[2:5:1])

# Use enumerate method to get index and values in the list
for index, value in enumerate(countries):
    print(f"index = {index}, value = {value}")
```

  **a**. Given the two lists of same length, create a dictionary using a list as keys and another list as values without using loop<br>Eg: ["a","b","c","d"],[1,2,3,4] lists should be made {'a': 1, 'b': 2, 'c': 3, 'd': 4}<br>Hint: zip method<br>
  **b**. Create two dictionaries, merge them into one.<br>Hint: update method/dictionary unpacking<br>
  **c**. Given the list of students, assign them the course.<br>Students=[“Duy”, “Laxmi”,”Antonio”,”Maria”]<br>Course=["Python","Database","ML"] <br>Should result to the new dictionary students_profile=<br>{'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'],<br> 'Antonio': ['Python', 'Database', 'ML'],'Maria': ['Python', 'Database', 'ML']}<br>

  **f**. Replace Antonio’s “ReactJs” to “Vuejs”<br>
  **g**. Remove an item from dictionary<br>Hint: use del statement or pop/ popitem method. Observe the differences<br>


```python
# Given the two lists of same length, create a dictionary using a list as keys and another list as values without using loop
#  Eg: ["a","b","c","d"],[1,2,3,4] lists should be made {'a': 1, 'b': 2, 'c': 3, 'd': 4}
#  Hint: zip method

dict_1 = dict(zip(['a', 'b', 'c', 'd'], [1, 2, 3, 4]))
dict_2 = dict(zip(['e', 'f', 'g'], [5, 6, 7]))
print("dict_1 = " + str(dict_1))
print("dict_2 = " + str(dict_2))

# Create two dictionaries, merge them into one.
#  Hint: update method/dictionary unpacking

# Solution 1 (probably naive): get keys and values as lists -> merge the lists -> then use the same zip method as above on the combined lists
dict_12 = dict(zip(list(dict_1.keys()) + list(dict_2.keys()), list(dict_1.values()) + list(dict_2.values())))
print("dict_12 = " + str(dict_12))

# Solution 2 (after reading up on the update method): 
dict_12 = dict_1
dict_12.update(dict_2)
print("dict_12 = " + str(dict_12))

# Given the list of students, assign them the course.
#  Students=[“Duy”, “Laxmi”,”Antonio”,”Maria”]
#  Course["Python","Database","ML"] 
# Should result to the new dictionary students_profile = 
#  {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'],
#   'Antonio': ['Python', 'Database', 'ML'],'Maria': ['Python', 'Database', 'ML']}
#
# My solution: we will use the dict(zip(...,...)) method again; just need to turn the "courses" list into a list of lists
students = ["Duy", "Laxmi", "Antonio", "Maria"]
courses  = ["Python", "Database", "ML"] 
dict_3 = dict(zip(students, [courses]*len(students)))
print("dict_3 = " + str(dict_3))

# Modify Antonio’s course to ["JavaScript","Database","NodeJS"]
dict_3['Antonio'] = ["JavaScript","Database","NodeJS"]
print("dict_3 = " + str(dict_3))

# Add “ReactJs” to Anonio’s course
dict_3['Antonio'].append('ReactJs')
print("dict_3 = " + str(dict_3))

# Replace Antonio’s “ReactJs” to “Vuejs”
# My solution: use list comprehension (there might be a simpler way?)
dict_3['Antonio'] = ['Vuejs' if s == 'ReactJs' else s for s in dict_3['Antonio']]
print("dict_3 = " + str(dict_3))

# Remove an item from dictionary
#  Hint: use del statement or pop/popitem method. Observe the differences
#  My solution: first create (deep) copies so that the del and pop are not performed on the same object; then perform the two deletion option and observe the difference
import copy
dict_4a = copy.deepcopy(dict_3)
dict_4b = copy.deepcopy(dict_3)
dict_4a.pop('Laxmi')
del dict_4b['Laxmi']
print("dict_4a = " + str(dict_4a))
print("dict_4b = " + str(dict_4b))
# The two removal methods have identical effects on the dictionary. The difference is that pop returns
# the removed element, while del doesn't (function vs procedure)
```
