**Exercise 1.** Iterator and generator: (For advanced learners, read from the links ) a. As you have already known, the iterables objects are the stream of data like string,list, dict, set etc which can be looped over or iterated. Since you have also known that you can convert an iterable into an iterator using iter method and iterators have next method, it's time to play around with the builtin itertools module.

i. Create a list apply iter method to it and run next method until you get StopIteration error

```python
# Create a list 
a_list = ['hello', 'this', 'is', 'a', 'list']

# Apply iter method to it 
a_list_iter = iter(a_list)

# Run next method until you get StopIteration error
while True:
    try:
        print(next(a_list_iter))
    except StopIteration:
        print('StopIteration error')
        break
```

**Exercise ii.** Import itertools in your notebook and create infinite iterators using count,cycle and and repeat methods. Refer https://docs.python.org/3/library/itertools.html

```python
import itertools

# create infinite iterators
count_it_15_step = itertools.count(10,15)
cycle_it = itertools.cycle('rotato')
repeat_it = itertools.repeat('again')

# create list with the first 12 next returns of each iterator
count_it_list = []
cycle_it_list = []
repeat_it_list = []

for _ in range(0,13):    
    count_it_list.append(next(count_it_15_step))
    cycle_it_list.append(next(cycle_it))
    repeat_it_list.append(next(repeat_it))
    
# print results
print('count_it_list = ' + str(count_it_list) + '\n')
print('cycle_it_list = ' + str(cycle_it_list) + '\n')
print('repeat_it_list = ' + str(repeat_it_list) + '\n')
```

iii. Use methods from I terators terminating on the shortest input sequence

```python
# create two lists of unequal lengths
list_short = list(range(0,3))
list_long = list(range(0,5))

print("list_short = " + str(list_short))
print("list_long = " + str(list_long))
    
# play around with some of the iterors
print("\n---zip_longest---")
result = itertools.zip_longest(list_short, list_long)
for x in result:
    print(str(x))
    
print("\n---chain---")
result = itertools.chain(list_short, list_long)
for x in result:
    print(str(x))
    
print("\n---filterfalse---")
result = itertools.filterfalse(lambda x: x in [3, 1, 4], list_long)
for x in result:
    print(str(x))
    
print("\n---dropwhile---")
result = itertools.dropwhile(lambda x: x < 2, list_long)
for x in result:
    print(str(x))
```

**Exercise iv.** User methods from Combinatoric iterator b. Create an infinite generator that generates the fibonacci series<br>

```python
# Combinatoric iterator example
perm_it = itertools.permutations(['fun', 'with', 'iterators'])
while True:
    try:
        print(' '.join(next(perm_it)))
    except:
        print("--the end--")
        break

# Create an infinite generator that generates the fibonacci series
print('\nFibonacci:')
def fib():
    curr_val = 0
    next_val = 1
    while True:
        yield curr_val
        tmp = curr_val
        curr_val = next_val
        next_val = tmp + next_val
        
fib_it = fib()

for _ in range(0,10):
    print(str(next(fib_it)))
```

**Exercise a.** Regular expression:

Read and practices more on regex

```python
import re

s = 'your  base are belong to us' 

# A regular expression to match the word 'base'
match = re.search(r'base', s)
print('start index:', match.start())
print('end index:', match.end())

# A regular expression to match the first three words
pattern = r'^\w+\s+\w+\s+\w+'  # ^: start of string; \w+: match one or more characters (i.e., a word); \s+: match one or more spaces
first_words = re.findall(pattern, s)
print(first_words)

# A regular expression to match the last three words in the given string.
pattern = r'\w+\s+\w+\s+\w+$'  # \w+: match one or more characters (i.e., a word); \s+: match one or more spaces; $: end of string
last_words = re.findall(pattern, s)
print(last_words)

# A regular expression to match all words that start with the letter 'b' 
pattern = r'b\w*'  
b_words = re.findall(pattern, s)
print(b_words)

# A regular expression to match all words that end with the letter 's' 
pattern = r'\b\w+s\b'
s_words = re.findall(pattern, s)
print(s_words)
```

**Exercise ii.** Write a multiplication table of any given number in a list using list comprehension

```python
def multi_table(m, n = 10):
    """ Returns the first n elements of the multiplication table for number m """
    return [m*x for x in range(1,n+1)]

print(str(multi_table(3,10)))
```

**Exercise viii.** Run the following command in your notebook. What could be the reason the list comprehension is more elegant?



**Answer**<br>
The solution with list comprehension is more elegant for two reasons:
 1. It is more concise, which helps to keep the code more readable. 
 2. It is faster than an ordinary loop due some optimizations built into the list comprehension function.

```python
import time

# Using for loop
start_time = time.time()

nums = []
for i in range(1, 10000001):
    nums.append(i**2)

end_time = time.time()
for_loop_time = end_time - start_time

# using list comprehension
start_time = time.time()

nums = [i**2 for i in range(1,10000001)]

end_time = time.time()
list_comp_time = end_time - start_time

print("For loop time:", for_loop_time)
print("List comprehension time:", list_comp_time)
```
