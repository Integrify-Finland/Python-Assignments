### Function to check if number is Prime

```python
def is_prime(n: int):
    """Function to check if a number is prime"""
    
    if n<=1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        #print(i)
        if n % i == 0:
            return False
    return True
```

```python
is_prime(100)
```

```python
def is_prime_list(n: int):
    """Function to check if a number in a list is prime"""
    
    val = []
    for i in (n):
        print(i)
        if i<=1:
            val.append('False')
            #return val
        for j in range(2, int(i ** 0.5) + 1):
            # print(j)
            if i % j == 0:
                val.append('False')
                #return val
            else:
                val.append('True')
                return val
    return val
```

### Unit Test for is_prime() function

```python
import unittest

class TestIsPrime(unittest.TestCase):
    def test_neg_prime(self):
        assert is_prime(-3), "Negative values are not prime numbers"
        
    def test_zero_and_one_prime(self):
        assert is_prime(0), "The value 0 and 1 is not a prime number"
        
    def test_small_prime(self):
        assert is_prime(3), "The small value is not a prime number"
        
    def test_small_composite_prime(self):
        assert is_prime(8), "The small composite value is not a prime number"
        
    def test_large_prime(self):
        assert is_prime(107), "The large value is not a prime number"
        
    def test_large_composite_prime(self):
        assert is_prime(200), "The large composite value is not a prime number"

unittest.main(argv=[''], verbosity=2, exit=False)
```
