```python
# 1. Use the **constructor methods** to create different data types in python 
# (integer,float,string,list,tuple,dictionary,set etc.). 
# Also use **object literals** to create those data types.

# Integer
print("Integer:")
# Constructor method
x_c_int = int(5)
print(f"{x_c_int} using constructor method.\n") 

# Object literals
x_o_int = 5
print(f"{x_o_int} using object literals.\n")

# Float
print("Float:")
x_c_float = float(5.25)
x_o_float = 5.75
print(f"{x_c_float} is in constructor method and {x_o_float} is an object literal.\n")

# String
print("String:")
name_c_str = str("Reethika")
name_o_str = "Ambatipudi"
print(f"{name_c_str} is in constructor method and {name_o_str} is an object literal.\n")

# List
print("List:")
x_c_list = list([1, 2, 3])
x_o_list = [1,2,3]
print(f"{x_c_list} is in constructor method and {x_o_list} is an object literal.\n")

# Tuple
print("Tuple:")
x_c_tuple = tuple((1,2,3))
x_o_tuple = (1,2,3)
print(f"{x_c_tuple} is in constructor method and {x_o_tuple} is an object literal.\n")

# Dictionary
print("Dictionary:")
x_c_dict = dict(week ='Saturday', day = 11, month = 3, year = 2023)
x_o_dict = {'week' :'Saturday', 'day' : 11, 'month' : 3, 'year' : 2023}
print(f"{x_c_dict} is in constructor method and {x_o_dict} is an object literal.\n")

# Set
print("Set:")
x_c_set = set([1, 1,  2, 3 , 4, 0, 2])
x_o_set = {1, 1,  2, 3 , 4, 0, 2}
print(f"{x_c_set} is in constructor method and {x_o_set} is an object literal.\n")
```

```python
# 2. Operators in python
# (https://www.w3schools.com/python/python_operators.asp)
# a. Create two operands and play around using different operators.
# b. Once you create an object of each data type, run **dir(object)**. 
# Navigate the **magic methods/dunder methods** 
# and other methods and attributes of each object. 
# Try to implement those methods and attributes as many as possible.


a = 10
b = 5
print(f"a = {a} and b = {b}\n")

# Arithmetic operators
print("Arithmetic operators:")
print(f"a+b = {a + b}")  
print(f"a-b = {a - b}") 
print(f"a*b = {a * b}")  
print(f"a/b = {a / b}") 
# Floor division rounds the result down to the nearest whole number
print(f"a//b = {a // b}") 
# % gives reminder and ** gives exponent
print(f"a%b = {a % b}")   
print(f"a**b = {a ** b}\n") 


# Comparison operators
print("Comparison operators:")
print(f"a==b -> {a == b}")  
print(f"a!=b -> {a != b}")  
print(f"a>b -> {a > b}")   
print(f"a<b -> {a < b}")  
print(f"a>=b -> {a >= b}") 
print(f"a<=b -> {a <= b}\n") 


# Bitwise operators
print("Bitwise operators:")
print(a & b)   
print(a | b) 
# XOR	Sets each bit to 1 if only one of two bits is 1
print(a ^ b) 
# NOT	Inverts all the bits
print(~a)  
# Zero fill left shift	Shift left by pushing zeros in from the right and let the leftmost bits fall off
print(a << 1)  
# Signed right shift	Shift right by pushing copies of the leftmost bit in from the left, 
# and let the rightmost bits fall off
print(f"{a >> 1} \n")  

# Assignment Operators
print("Assignment Operators:")
x=5
y=10
z = x
z += y
print(z) 

z = x
z -= y
print(z)

z = x
z *= y
print(z) 

z = x
z /= y
print(z) 

z = x
z //= y
print(z) 

z = x
z %= y
print(z) 

z = x
z **= y
print(f"{z}\n") 


# Logical Operators
print("Logical Operators:")
a = True
b = False
print(a and b)
print(a or b)
print(f"{not a}\n") 


# Identity Operators
print("Identity Operators:")
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a is b) 
print(a is not b)
print(a is c) 
print(f"{a == b}\n") 


# Membership Operators
print("Membership Operators:")
a = [1, 2, 3]
print(1 in a) 
print(f"{4 not in a}\n") 
```

```python
# Learn different ways to format the string.
# Tutorials on (https://realpython.com/python-formatted-output/)

print("I love", "Python")
print("I love" + " Python")
print("I love " + "Python " + "version" , 3.11)
print("I love"+" python" + "version %.2f"%3.11)
print("I love %s version %.2f"%("python", 3.11)) 

# Using format positional argument
print("I love {} version {}".format("python", 3.11))
print("I love {} version {}".format("python", "3.11"))
print("I love {} version {:.2f}".format("python",3.11)) 
print("I love {1} version of {0}".format("python", "3.11"))

# Using format keyword argument
print("I love {pl} version {version:.21}". format(pl="python", version=3.11))

# Using f string literal
info={"pl": "python","version":3.11}
print(f"I love {info.get('pl')} version {info.get('version'):.2f}")
    
# Can be done with uppercase Literal and single quote but double quote
# while accessing dictionary value 
print(F'I love {info["pl"]} version {info["version"]}')

batch = "ML"
print(f"I love {'javascript' if batch=='FS' else 'python'}")
```

```python
print("Familiarize yourself with indexing in python:")

# Define the list of countries
countries= ["Finland","Sweden","Denmark","Netherlands","Germany","Norway","Iceland"]

# 1. Find the index of "Sweden".
print(countries.index("Sweden"))  

# 2. Get all the items except the first
print(countries[1:])  

# 3. Get all the items except the last
print(countries[:-1])  

# 4. Find the last item using negative indexing
print(countries[-1]) 

# 5. Find the last three item using negative indexing
print(countries[-3:])  

# 6. Grab all the items with odd indices
print(countries[1::2]) 

# 7. Find the first and last item for the list
print(countries[0], countries[-1])  

# 8. Get ["Netherlands","Germany","Norway"] using negative indexing
print(countries[-4:-1])  

# 9. Reverse the list
countries.reverse()
print(countries)  

# 10. What is the output of countries[::-2]?
print(countries[::-2])  

# 11. What is the output of countries[1:100:2]?
# The output will be the same as countries[1::2] since there are only 7 items in the list
print(countries[1:100:2])  

# 12. What is the output of countries[len(countries)-1:0:2]?
# This will result in an empty list since the starting index is after the ending index
print(countries[len(countries)-1:0:2]) 

# 13. Get the output ['Germany', 'Netherlands', 'Denmark']
print(countries[-4:-7:-1])  

# 14. Use enumerate method to get index and values in the list
for index, value in enumerate(countries):
    print(f"{index}: {value}")
```

```python
print("Exercise on dictionaries")

list1 = ["a","b","c","d"]
list2 = [1,2,3,4]
result_dict = dict(zip(list1, list2))
print(result_dict)


dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
dict1.update(dict2)
print(dict1)

Students=["Duy", "Laxmi","Antonio","Maria"]
Course=["Python","Database","ML"]
students_profile = {student: Course for student in Students}
print(students_profile)


students_profile['Antonio'] = ["JavaScript", "Database", "NodeJS"]
print(students_profile)

students_profile['Antonio'].append("ReactJs")
print(students_profile)

students_profile['Antonio'][2] = "Vuejs"
print(students_profile)


del students_profile['Maria']
print(students_profile)

students_profile.pop("Antonio")
print(students_profile)

students_profile.popitem()
print(students_profile)

# pop() method can be used to remove a specific item from the dictionary based on the key. 
# popitem() method can be used to remove and return the last inserted item (key-value pair) from the dictionary.
```
