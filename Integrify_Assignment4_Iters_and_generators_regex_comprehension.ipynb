### 1.4 Use methods from Combinatoric iterator
### Combinatoric iterators

```python
"""
Product(): This tool computes the cartesian product of input iterables. 
To compute the product of an iterable with itself, we use the optional repeat 
keyword argument to specify the number of repetitions. 
The output of this function is tuples in sorted order.
"""

# import the product function from itertools module
from itertools import product

print("The cartesian product using repeat:")
print(list(product([1, 2], repeat = 2)))
print()

print("The cartesian product of the containers:")
print(list(product(['geeks', 'for', 'geeks'], '2')))
print()

print("The cartesian product of the containers:")
print(list(product('AB', [3, 4])))
```

```python
"""
Permutations(): Permutations() as the name speaks for itself is used 
to generate all possible permutations of an iterable. 
All elements are treated as unique based on their position and not their values. 
This function takes an iterable and group_size, if the value of group_size is not specified or 
is equal to None then the value of group_size becomes the length of the iterable.
"""
# import the product function from itertools module
from itertools import permutations

print ("All the permutations of the given list is:")
print (list(permutations([1, 'integrify'], 2)))
print()

print ("All the permutations of the given string is:")
print (list(permutations('AB')))
print()

print ("All the permutations of the given container is:")
print(list(permutations(range(3), 2)))
```

```python
"""
Combinations(): This iterator prints all the possible combinations (without replacement) 
of the container passed in arguments in the specified group size in sorted order.
"""

# import combinations from itertools module

from itertools import combinations

print ("All the combination of list in sorted order(without replacement) is:")
print(list(combinations(['A', 2], 2)))
print()

print ("All the combination of string in sorted order(without replacement) is:")
print(list(combinations('AB', 2)))
print()

print ("All the combination of list in sorted order(without replacement) is:")
print(list(combinations(range(2), 1)))
```

```python
"""
Combinations_with_replacement(): This function returns a subsequence of length n 
from the elements of the iterable where n is the argument that the function takes 
determining the length of the subsequences generated by the function. 
Individual elements may repeat itself in combinations_with_replacement function.
"""

# import combinations from itertools module

from itertools import combinations_with_replacement

print ("All the combination of string in sorted order(with replacement) is:")
print(list(combinations_with_replacement("AB", 2)))
print()

print ("All the combination of list in sorted order(with replacement) is:")
print(list(combinations_with_replacement([1, 2], 2)))
print()

print ("All the combination of container in sorted order(with replacement) is:")
print(list(combinations_with_replacement(range(2), 1)))
```

## Create an infinite generator that generates the fibonacci series
Reference: https://stackoverflow.com/questions/3953749/python-fibonacci-generator

```python
import itertools as it

num_iterations = int(input('How many? '))
def getFibonacci():
    a,b = 0,1
    while True:
        yield a
        b = a+b
        yield b
        a = a+b

fib_list=[]
for x in it.islice(getFibonacci(), num_iterations):
    fib_list.append(x)
print(fib_list)
```

### 1.2 Write a multiplication table of any given number in a list using list comprehension

```python
def multiplication_table(num, num_s):
    
    print('Multiplication table for:', num)
    for i in range(1, num_s + 1):
        a = i
        product = num*i
        print(num, "*", a, " = ", product)
    #return None

# Run the function
multiplication_list = [1, 2, 3, 4, 5, 6, 7]

list_using_comp = [multiplication_table(var, 12) for var in multiplication_list  if var % 2 == 0]

print(list_using_comp)
```

### 1.8 Run the following command in your notebook. What could be the reason the list comprehension is more elegant?



```python
import time

# Using for loop
start_time = time.time()

nums = []
for i in range(1, 10000001):
    nums.append(i**2)
    
end_time = time.time()
for_loop_time = end_time - start_time

# Using list comprehension
start_time = time.time()

nums = [i**2 for i in range(1, 10000001)]

end_time = time.time()
list_comp_time = end_time - start_time

print("For loop time: ", for_loop_time)
print("List comprehension time: ", list_comp_time)
```

### Is list comprehension better?
Reference: https://towardsdatascience.com/list-comprehensions-vs-for-loops-it-is-not-what-you-think-34071d4d8207

### List comprehensions are faster than For loops to create lists.

1. This is because in the "For Loop", we are creating a list by appending new elements to it at each iteration. This is slow.

2. List comprehensions are also more declarative than loops, which means they're easier to read and understand. 

3. For Loops require you to focus on how the list is created. You have to manually create an empty list, loop over the elements, and add each of them to the end of the list.

### For Loops are Faster than List Comprehensions
Suppose we only want to perform some computations (or call an independent function multiple times) and do not want to create a list.

In that case, we see that the list comprehension is **slower** than the for loop.

```python
import time
iterations = 100000000

start = time.time()
for i in range(iterations):
    i+1
end = time.time()
print(end - start)
#6.16 seconds

start = time.time()
[i+1 for i in range(iterations)]
end = time.time()
print(end - start)
#7.80 seconds
```
