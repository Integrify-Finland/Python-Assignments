Distance Metrics are important to calculate the similarity between the vectors. They are used in both supervised and unsupervised machine learning problems. Given the points (1,2,3) and (4,5,6), write a python conditional statement to print the distance for the order 1,2,3,4. If the order is 1 then it is L1 norm(Manhattan distance), order 2 is L2 norm(Euclidean distance) and the generalized distance is given by Minkowski equation.

```python
# Define the two points
point1 = (1, 2, 3)
point2 = (4, 5, 6)

# Define the order
order = 3

# Calculate the distance based on the order
if order == 1:
    # L1 norm (Manhattan distance)
    distance = sum(abs(p1 - p2) for p1, p2 in zip(point1, point2))
elif order == 2:
    # L2 norm (Euclidean distance)
    distance = (sum((p1 - p2)**2 for p1, p2 in zip(point1, point2)))**0.5
else:
    # Generalized distance (Minkowski equation)
    distance = (sum(abs(p1 - p2)**order for p1, p2 in zip(point1, point2)))**(1/order)

# Print the calculated distance
print(f"The distance between {point1} and {point2} for order {order} is {distance:.2f}.")
```

```python
#using numpy array
import numpy as np

# Define the two points as NumPy arrays
point1 = np.array([1, 2, 3])
point2 = np.array([4, 5, 6])

# Define the order
order = 2

# Calculate the distance based on the order
if order == 1:
    # L1 norm (Manhattan distance)
    distance = np.sum(np.abs(point1 - point2))
elif order == 2:
    # L2 norm (Euclidean distance)
    distance = np.linalg.norm(point1 - point2)
else:
    # Generalized distance (Minkowski equation)
    distance = np.power(np.sum(np.power(np.abs(point1 - point2), order)), 1.0 / order)

# Print the calculated distance
print(f"The distance between {point1} and {point2} for order {order} is {distance:.2f}.")
```

```python
#Using dictionary mapper
# Given template string
dna_template = 'AATCCGAAAATTCGGGAATTTTCGCGT'

# Define the mapper dictionary
mapper = {"T":"A", "A":"T", "G":"C", "C":"G"}

# Generate the complementary DNA template
complementary_dna = ''
for base in dna_template:
    complementary_base = mapper[base]
    complementary_dna += complementary_base

# Print the result
print(complementary_dna)
```

#8.
Generate a list of n numbers of prime numbers

**Brute-force** approach is a straightforward method of solving a problem by trying all possible solutions, without necessarily optimizing the process. It involves checking every possibility systematically and exhaustively until a solution is found. T*his approach can be useful for small problems, but can quickly become impractical or even impossible for larger ones*.

In the context of programming, brute-force algorithms are often used as a starting point for solving more complex problems. For example, when searching for a particular value in a list, a brute-force approach would involve checking each element of the list until the desired value is found. Similarly, when generating prime numbers, a brute-force approach would involve checking each candidate number for primality by dividing it by every smaller number, until a divisor is found.

Although brute-force algorithms are simple to understand and implement, they can be very inefficient for large problem sizes, since they may involve checking a huge number of possibilities. Therefore, more sophisticated algorithms are often needed to solve larger and more complex problems efficiently.
â€‹




```python
#using a brute-force approach
def generate_primes(n):
    primes = []  # Initialize an empty list to store primes
    num = 2     # Start checking for primes from 2
    while len(primes) < n:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
        num += 1
    return primes

primes = generate_primes(10)
print(primes)
```

```python
#using the Sieve of Eratosthenes algorithm
def generate_primes2(n):
    if n < 2:
        return []
    primes = [2]  # Initialize the list of primes with 2
    sieve = [True] * (n + 1)  # Initialize the sieve with True values
    for i in range(3, n + 1, 2):  # Only check odd numbers
        if sieve[i]:
            primes.append(i)
            for j in range(i * i, n + 1, 2 * i):
                sieve[j] = False
    return primes

primes = generate_primes2(30)
print(primes)
```

Perform **the minmax normalization** to the range [0,1] for the data1 and minmax normalization to the range [-1,1] for the data2
data1=[1,17,12,15,30,60,48,51,89,92,54,8], use

X_scaled = (X - X_min) / (X_max - X_min)

data2=[-13,2,-1,7,22,-32,5,15,-9,13,3,-2], use

X_scaled = 2 * (X - X_min) / (X_max - X_min) - 1

The general formula for scaling in the range [a,b] is

X_scaled = (b-a) * (X - X_min) / (X_max - X_min) + a



```python
def minmax_scale_01(data):
    """
    Perform min-max normalization to the range [0,1] for a given data.
    
    Parameters:
    data (list): The input data to be normalized.
    
    Returns:
    list: The normalized data.
    """
    X_min = min(data)
    X_max = max(data)
    X_scaled = [(x - X_min) / (X_max - X_min) for x in data]
    return X_scaled

data1 = [1, 17, 12, 15, 30, 60, 48, 51, 89, 92, 54, 8]
data1_scaled = minmax_scale_01(data1)
print(data1_scaled)
```

```python
def minmax_scale_11(data):
    """
    Perform min-max normalization to the range [-1,1] for a given data.
    
    Parameters:
    data (list): The input data to be normalized.
    
    Returns:
    list: The normalized data.
    """
    X_min = min(data)
    X_max = max(data)
    X_scaled = [2 * (x - X_min) / (X_max - X_min) - 1 for x in data]
    return X_scaled

data2 = [-13, 2, -1, 7, 22, -32, 5, 15, -9, 13, 3, -2]
data2_scaled = minmax_scale_11(data2)
print(data2_scaled)
```

Understand how sorting algorithms work(bubble,selection,insertion). Implement them of your own. Also compare time complexities.

**Bubble Sort**
 is a simple sorting algorithm that repeatedly iterates through a list, *comparing adjacent elements and swapping them *if they are in the wrong order. The algorithm continues until no more swaps are needed, indicating that the list is sorted.

he time complexity of bubble sort is **O(n^2)** in the worst case, where n is the length of the list. This is because the algorithm may need to make up to n-1 passes through the list, and on each pass it may need to compare up to n-1 adjacent elements.

```python
def bubble_sort(arr):
    """Sorts a list of integers using the bubble sort algorithm.
    
    Args:
        arr (list): The list of integers to be sorted.
    """
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

**Selection Sort**
 is another simple sorting algorithm that works by repeatedly finding the minimum element from the unsorted part of a list and swapping it with the first element of the unsorted part. The algorithm continues until the entire list is sorted.

The time complexity of selection sort is also** O(n^2)** in the worst case, since it may need to make up to n-1 passes through the list, and on each pass it may need to find the minimum element by comparing up to n-1 elements.

```python
def selection_sort(arr):
    """Sorts a list of integers using the selection sort algorithm.
    
    Args:
        arr (list): The list of integers to be sorted.
    """
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
```

**Insertion Sort**
 is a sorting algorithm that works by iteratively inserting elements from an unsorted list into their correct positions in a sorted list. The algorithm builds the sorted list one element at a time, until the entire unsorted list is sorted.

```python
def insertion_sort(arr):
    """Sorts a list of integers using the insertion sort algorithm.
    
    Args:
        arr (list): The list of integers to be sorted.
    """
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
```

#12.
Write a function factorial of a given number

```python
def factorial(n):
    """
    Calculate the factorial of a given number.
    
    Parameters:
    n (int): The input number to calculate the factorial of.
    
    Returns:
    int: The factorial of the input number.
    """
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
# Example usage
n = 5
fact = factorial(n)
print(f"The factorial of {n} is {fact}")
```

#13.
Write a function that takes two arguments: activation_function name and value that returns the output of specific activation_function. Implement for sigmoid, binary step function, relu, leaky rely, tanh.

Eg : activation_function('sigmoid',0) should return 0.5

```python
import math

def activation_function(name, x):
    """
    Compute the output of an activation function for a given input value.
    
    Parameters:
    name (str): The name of the activation function to use. Supported functions: 'sigmoid', 'binary_step', 'relu', 'leaky_relu', 'tanh'.
    x (float): The input value to the activation function.
    
    Returns:
    float: The output of the activation function for the input value.
    """
    if name == 'sigmoid':
        return 1 / (1 + math.exp(-x))
    elif name == 'binary_step':
        return 1 if x >= 0 else 0
    elif name == 'relu':
        return max(0, x)
    elif name == 'leaky_relu':
        return max(0.1 * x, x)
    elif name == 'tanh':
        return math.tanh(x)
    else:
        raise ValueError(f"Unsupported activation function '{name}'.")
```

```python
# Example usage
x = 0
sigmoid_output = activation_function('sigmoid', x)
print(f"The sigmoid output for x={x} is {sigmoid_output}")

binary_step_output = activation_function('binary_step', x)
print(f"The sigmoid output for x={x} is {sigmoid_output}")
```

```python
# Example usage
x = 0.25
sigmoid_output = activation_function('sigmoid', x)
print(f"The sigmoid output for x={x} is {sigmoid_output}")

binary_step_output = activation_function('binary_step', x)
print(f"The sigmoid output for x={x} is {sigmoid_output}")
```

```python
# Example 1: Sigmoid activation function
output = activation_function('sigmoid', 0)
print(output) # Output: 0.5

# Example 2: Binary step activation function
output = activation_function('binary_step', 5)
print(output) # Output: 1

# Example 3: ReLU activation function
output = activation_function('relu', -3)
print(output) # Output: 0

# Example 4: Leaky ReLU activation function
output = activation_function('leaky_relu', -3)
print(output) # Output: -0.03

# Example 5: Tanh activation function
output = activation_function('tanh', 1)
print(output) # Output: 0.7615941559557649
```

Write a function that takes three arguments. First arguments should determine whether to return permutation or combination or both. Make the first argument optional and default is both if nothing is passed while calling the function.The second argument is n, which is optional and default is 10. Finally the third argument is r, which is also optional and default is 5.

This function uses the itertools library to generate permutations and combinations of integers from 1 to n, taken r at a time. The first argument p_or_c determines whether to return "permutation", "combination", or "both". If "permutation" is chosen, the function returns a list of permutations. If "combination" is chosen, the function returns a list of combinations. If "both" is chosen (or no argument is provided), the function returns a dictionary with two keys, "permutations" and "combinations", each with a list of the corresponding values

```python
import itertools

def perm_or_comb(perm_or_comb="both", n=10, r=5):
    if perm_or_comb == "both":
        return (list(itertools.permutations(range(n), r)), list(itertools.combinations(range(n), r)))
    elif perm_or_comb == "permutation":
        return list(itertools.permutations(range(n), r))
    elif perm_or_comb == "combination":
        return list(itertools.combinations(range(n), r))
    else:
        raise ValueError("Invalid perm_or_comb argument value.")
```

#16.
Implement a function that takes any number of positional and keyword arguments.

In Python, you can define a function that takes any number of positional and keyword arguments using the *args and **kwargs syntax.

The *args parameter allows the function to accept any number of positional arguments as a tuple, while the **kwargs parameter allows the function to accept any number of keyword arguments as a dictionary.

 *args will capture all positional arguments passed to the function and store them in a tuple named args. Similarly, **kwargs will capture all keyword arguments passed to the function and store them in a dictionary named kwargs.

```python
def example_function(*args, **kwargs):
    print("Positional arguments:")
    for arg in args:
        print(arg)

    print("\nKeyword arguments:")
    for key, value in kwargs.items():
        print(key + " = " + str(value))
```

Understand scope of variables: LEGB Rule 1. Local to the function eg. variable defined in the function 2. Enclosing or nonlocal eg. variables defined in enclosing function in nested function 3. Global eg. variables defined within main program or module 4. Built-in eg. Python keywords
Note: Names of global and built-in can be accessed from anywhere

Create an example of each case above to illustrate.

**Local Variables**
are variables that are defined inside a function and can only be accessed within that function.

```python
def example_function():
    x = 10    # Local variable
    print("Inside the function, x =", x)

example_function()
```

**Enclosing or Nonlocal Variables**
are variables that are defined in an enclosing function and can be accessed by nested functions within that enclosing function.

```python
def outer_function():
    x = 10    # Enclosing variable
    
    def inner_function():
        nonlocal x
        x = 20    # Modify the enclosing variable
        print("Inside the inner function, x =", x)
    
    inner_function()
    print("Inside the outer function, x =", x)

outer_function()
```

**Global variables** are variables that are defined outside of any function and can be accessed from anywhere in the code.

```python

def example_function():
    x=5
    print("Inside the function, x =", x)
x=10
example_function()
# Output: Inside the function, x = 10

# You can also modify the global variable from within a function:
def modify_global_variable():
    global x
    x = 20    # Modify the global variable
    print("Inside the function, x =", x)

modify_global_variable()
print("Outside the function, x =", x)
```
